```python
import pandas as pd
import matplotlib.pyplot as plt

aviation_data = pd.read_csv("Monthly_figures_on_aviation_11102023_104644.csv", delimiter=';')
gdp_data = pd.read_csv("Approaches_of_domestic_product__GDP__16102023_091908.csv", delimiter=';')

# Select the specific columns and rows
aviation_column = "Commercial air traffic/Flights/All flights/Total flights (number)"
aviation_data = aviation_data.loc[1:21, ["Periods", aviation_column]]

gdp_column = "GDP from the generation of income/Value at current prices/Gross domestic product (million euros)"
gdp_data = gdp_data.loc[0:20, ["Periods", gdp_column]]

# Convert "Periods" column to a datetime format (if it's not already)
aviation_data["Periods"] = pd.to_datetime(aviation_data["Periods"], format='%Y')
gdp_data["Periods"] = pd.to_datetime(gdp_data["Periods"], format='%Y')

plt.figure(figsize=(12, 6))

# Create a single subplot for both datasets
plt.plot(aviation_data["Periods"], aviation_data[aviation_column], label="Commercial Air Traffic")
plt.plot(gdp_data["Periods"], gdp_data[gdp_column], label="Gross domestic product")

plt.title("Commercial Air Traffic vs. Gross domestic product")
plt.xlabel("Year")
plt.ylabel("Value")
plt.legend()  # Add a legend to distinguish between the two datasets

plt.tight_layout()
plt.show()

print("Gross domestic product (GDP) is a quantity that expresses the size of an economy. The volume change of GDP during a reference period expresses the growth or shrinkage of the economy. Gross domestic product at market prices is the final result of the production activity of resident producer units. It can be defined in three ways:")
```
